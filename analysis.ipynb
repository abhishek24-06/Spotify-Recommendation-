```python
df=df[(df['tempo']>0) & (df['danceability']>0)]
```

Filter out Outlier based on Duration of tracks

```python
df=df[(df['duration_ms']<=900000) & (df['duration_ms']>=30000)]
```

Drop unnecessary Columns

```python
df=df.drop(columns=['id','name','artists','release_date','year','duration_ms','popularity'])
```

One hot encoding for Key Column

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
coln_to_encode=OneHotEncoder(handle_unknown='ignore',sparse_output=False)

encode_value=coln_to_encode.fit_transform(df[['key']])

new_coln=coln_to_encode.get_feature_names_out(['key'])

encode_df=pd.DataFrame(encode_value,columns=new_coln,index=df.index)
df=pd.concat([df,encode_df],axis=1)
```

```python
df=df.drop(columns=['key'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()

numeric_features = ['valence', 'acousticness', 'danceability', 
                    'energy', 'instrumentalness', 
                    'liveness', 'loudness', 'mode' , 'speechiness', 'tempo']

df[numeric_features]=scaler.fit_transform(df[numeric_features])
```

```python
X=df[numeric_features]
```

```python
from sklearn.neighbors import NearestNeighbors

clf=NearestNeighbors(n_neighbors=11,algorithm='brute',metric='cosine',n_jobs=-1)

clf.fit(X)

```

```python
df_original=pd.read_csv('data.csv')
```

```python
# Read original ONCE, clean it, then keep a reference copy
df_original = pd.read_csv('data.csv')

# Apply the SAME cleaning to both
df_original = df_original[(df_original['tempo'] > 0) & (df_original['danceability'] > 0)]
df_original = df_original[(df_original['duration_ms'] <= 900000) & (df_original['duration_ms'] >= 30000)]
df_original = df_original.reset_index(drop=True)  # ← KEY: reset indices

# Now build features from the SAME cleaned dataframe
# ... your scaling, encoding, KNN code ...

def recommend(song_name):
    matches = df_original[df_original['name'].str.lower() == song_name.lower()]
    if matches.empty:
        return "Song not found"
    idx = matches.index[0]
    distances, indices = clf.kneighbors([X.iloc[idx]])
    recommended_indices = indices[0][1:11]
    return df_original[['name', 'artists']].iloc[recommended_indices]
```
